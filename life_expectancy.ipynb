# Life Expectency Regression
This project uses the TensorFlow and Keras libraries

```python
# Library Import
import pandas as pd

import tensorflow as tf
import tensorflow.keras.layers as layers
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Data Preparation
dataset = pd.read_csv('/content/life_expectancy_data/life_expectancy.csv')
dataset.head()
```

```python
dataset.describe()
```

```python
dataset = dataset.drop(['Country'], axis=1)
dataset
```

```python
# Isolate labels from features
labels = dataset.iloc[:, -1]
labels
```

```python
# Isolcate features from labels
features = dataset.iloc[:, :-1]
features
```

## Data Preprocessing

```python
# Perform one-hot encoding of categorical features
features = pd.get_dummies(features)
features
```

```python
# Split feature set into training and testing data
features_train, features_test, labels_train, labels_test = train_test_split(
    features,
    labels,
    test_size=0.2,
    random_state=42)
```

```python
# Scale numeric values to prevent scale from exerting a bias on predited weights
numerical_features = features.select_dtypes(include=['int64', 'float64'])
numerical_columns = numerical_features.columns
ct = ColumnTransformer(
    [('only_numeric', StandardScaler(), numerical_columns)],
    remainder='passthrough'
)

# Fit the column transformer to the feature test and training data
features_train_scaled = ct.fit_transform(features_train)
features_test_scaled = ct.transform(features_test)
```

# Build the model

```python
# Create a model object and layer objects
my_model = Sequential()

input = layers.InputLayer(input_shape=(features.shape[1], ))
hidden = layers.Dense(64, activation="relu")
output = layers.Dense(1)

# Add layers to the model
my_model.add(input)
my_model.add(hidden)
my_model.add(output)

# Display output
print(my_model.summary())
```

```python
# Initialize Optimizer and Compile Model
adam = Adam(learning_rate=0.1)

my_model.compile(
    loss='mse',
    metrics=['mae'],
    optimizer=adam
)
```

# Fit and Evaluate Model

```python
my_model.fit(
    features_train_scaled,
    labels_train,
    epochs=40,
    batch_size=1,
    verbose=1
)
```

```python
res_mse, res_mae = my_model.evaluate(
    features_test_scaled,
    labels_test,
    verbose=0
)
print("Final loss:", res_mse)
print("Final metric:", res_mae)
```
